```python
import pandas as pd
import numpy as np
import re
```

## data

```python
df = pd.read_csv("../../data/pre/preprocessed1.csv")
```

```python
import re

splitted_words = df["상품명"].apply(lambda x:x.split(' '))
words_df = pd.DataFrame(word for word in splitted_words)
words_df.iloc[:,0] = words_df.iloc[:,0].apply(lambda x:re.sub('[()]','',x))
words_df.iloc[:,0] =  words_df.iloc[:,0].apply(lambda x: x.replace('[','').replace(']',''))
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
words_df2 = pd.merge(words_df, df["상품명"], left_index=True, right_index=True)
```

## 앞에 붙은 말 중 필요 없는 말 제거

```python
def remove_prefix(text, prefix):
    return text[text.startswith(prefix) and len(prefix):]

refined_words_list= []
for i, elems in enumerate(words_df2.iloc[:,0]):
    try:
        elem = remove_prefix(elems, "일시불")
        elem2 = remove_prefix(elem, "무이자")
        elem3 = remove_prefix(elem2, "일")
        elem4 = remove_prefix(elem3, "무")
        elem5 = remove_prefix(elem4, "ARS10%")
        elem6 = remove_prefix(elem5, "세일20%")
        words_df2.iloc[i,0] = elem6
    except:
        words_df2.iloc[i,0] = elems
```

## 필요없는 단어들 결측치 처리

```python
remove_list = ["일시불","무이자",'직매입','도냄무이자',
               '도냄일시불','일','무','2019','19','뉴','F/W','세일20%','가격인하','초특가','','"','1+1','1세트','19년','100%리얼',
               '2019년',"국내제조",'2019년형','2세트','5세트','1세트','4세트',"S/S","SS","국내생산","3인용","4인용","신제품","국내제작",
              "프랑스직수입"]
refined_df = words_df2.applymap(lambda x:x.replace(x, "NaN") if x in remove_list else x)
```

## 결측치 부분이 있으면 바로 옆 단어를 끌고 오도록

```python
for i in range(len(refined_df)):
    if ((refined_df.iloc[i,0] == "NaN") and (refined_df.iloc[i,1] == "NaN") and (refined_df.iloc[i,2] == "NaN")):
        refined_df.loc[i,"tmp_brand"] = refined_df.iloc[i,3]
    elif ((refined_df.iloc[i,0] == "NaN") and (refined_df.iloc[i,1] == "NaN")):
        refined_df.loc[i,"tmp_brand"] = refined_df.iloc[i,2]
    elif refined_df.iloc[i,0] == "NaN":
        refined_df.loc[i,"tmp_brand"] = refined_df.iloc[i,1]
    else:
        refined_df.loc[i,"tmp_brand"] = refined_df.iloc[i,0]
```
