```python
import numpy as np
import pandas as pd
import itertools
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Load dataset

```python
sales_train = pd.read_excel('sales_train.xlsx', header=1)
sales_test = pd.read_csv('sales_test_replace.csv')
```

```python
sales_train
```

```python
sales_test
```

```python
sales_train.nunique()
```

# divide 상품명 column by word

```python
word_pn_train = [x.split(' ') for x in sales_train['상품명'].tolist()]
```

```python
word_pn_train
```

# Make new columns from 상품명

## train data

```python
word_count_pn = list(itertools.chain(*word_pn_train))
word_count_pn = pd.Series(Counter(word_count_pn)).sort_values(ascending=False)
```

```python
pd.options.display.max_rows = 9999
word_count_pn
```

```python
word_count_pn.drop(labels=[''], inplace=True)
```

```python
print(len(word_count_pn[word_count_pn>1000]))
print(len(word_count_pn[word_count_pn>900]))
print(len(word_count_pn[word_count_pn>1100]))
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=word_count_pn.index, y=word_count_pn.values, data=word_count_pn[word_count_pn>1000])
```

```python
print(len(word_count_pn[word_count_pn>100]))
print(len(word_count_pn[word_count_pn>90]))
print(len(word_count_pn[word_count_pn>110]))
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=word_count_pn.index, y=word_count_pn.values, data=word_count_pn[word_count_pn>100])
```

```python
df_word_pn_train = pd.DataFrame(index=range(len(word_pn_train)), columns=word_count_pn[word_count_pn>100].index)
```

```python
df_word_pn_train
```

```python
for i in range(len(word_pn_train)):
    for j in range(len(word_pn_train[i])):
        for k in range(len(df_word_pn_train.columns)):
            if word_pn_train[i][j] == df_word_pn_train.columns[k]:
                df_word_pn_train.iloc[i,k] = 1
                
df_word_pn_train.fillna(0, inplace=True)
```

```python
sales_train_new = pd.concat([sales_train,df_word_pn_train], axis=1)
```

## test data

```python
word_pn_test = [x.split(' ') for x in sales_test['상품명'].tolist()]
```

```python
word_pn_test
```

```python
df_word_pn_test = pd.DataFrame(index=range(len(word_pn_test)), columns=word_count_pn[word_count_pn>100].index)
```

```python
df_word_pn_test
```

```python
for i in range(len(word_pn_test)):
    for j in range(len(word_pn_test[i])):
        for k in range(len(df_word_pn_test.columns)):
            if word_pn_test[i][j] == df_word_pn_test.columns[k]:
                df_word_pn_test.iloc[i,k] = 1
                
df_word_pn_test.fillna(0, inplace=True)
```

```python
sales_test_new = pd.concat([sales_test,df_word_pn_test], axis=1)
```

# Save new dataset

```python
sales_test_new.to_csv('sales_test_new.csv',index=False)
```

```python
sales_train_new.to_csv('sales_train_new.csv', index=False)
sales_test_new.to_csv('sales_test_new.csv', index=False)
```
